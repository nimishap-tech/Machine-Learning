# IMPORTING LIBRAIRIES

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as Sb
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler , LabelEncoder , OneHotEncoder
from sklearn.neighbors import KNeighborsClassifier 
from sklearn import metrics
from sklearn.metrics import accuracy_score , classification_report , confusion_matrix
from sklearn.neural_network import MLPClassifier
```

# EXPERIMENT NO1 :ANN

# DATASET1

```python
##Importing Data from the file 
df_data=pd.read_csv("C:/Users/PC/Desktop/ML/sgemm_product.csv")
df_data.head()

###Taking average log values for the Last four attributes.
df_data["Avg_run"]=df_data.iloc[:,14:].mean(axis=1)
df_data["log_avg_run"]=np.log(df_data["Avg_run"])
df_data.drop(['Run1 (ms)','Run2 (ms)','Run3 (ms)','Run4 (ms)','Avg_run'],axis=1,inplace=True)
df_data.head()

###Performing binary classification of data.
median=df_data.log_avg_run.median()
median
##Splitting data over median value.
X=df_data.iloc[:,1:14]
Y=df_data.iloc[:,14].apply(lambda x : 0 if x <= median else 1)
Y.value_counts()

###Performing standardization

X_s=X.copy()
standardscalar=StandardScaler()
X_s=standardscalar.fit_transform(X_s)

#########Spliting into Train-Test######################

X_train,X_test, y_train, y_test = train_test_split(X_s,Y,test_size = 0.25 , random_state = 0)
```

```python
#####Parameter1: Hidden Layers
from sklearn.neural_network import MLPClassifier
layer_size = np.arange(1,31,1)
acc_train_layers = []
acc_test_layers = []
for i in layer_size:
    ANN_class = MLPClassifier(hidden_layer_sizes= (i,i,i,i),
                          activation='relu', 
                          alpha=0.0001, 
                          batch_size='auto' ,
                          solver= 'adam',
                          learning_rate= 'constant',
                          max_iter = 300,
                          random_state= None )
    ANN_class.fit(X_train, y_train)
    y_pred_train = ANN_class.predict(X_train)
    y_pred_test = ANN_class.predict(X_test)
    acc_train_layers.append(accuracy_score(y_train,y_pred_train))
    acc_test_layers.append(accuracy_score(y_test,y_pred_test))
    
## Plotting the graph for the different hidden layer neurons 
plt.plot(layer_size,acc_train_layers,marker='o',markersize=6,color ='blue',label='training')
plt.plot(layer_size,acc_test_layers,marker='o',markersize=6,color='red',label='test')
plt.title(" Accuracies for different Hidden layers specified")
plt.xlabel("Three Hidden layer")
plt.grid()
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```

```python
## Parameter2:Activation Function
from sklearn.neural_network import MLPClassifier
function = ['identity', 'logistic', 'tanh', 'relu']
acc_train_func = []
acc_test_func = []
for i in function:
    ANN_class = MLPClassifier(hidden_layer_sizes= (30,30,30,30),
                          activation= i ,random_state= 0 )
    ANN_class.fit(X_train, y_train)
    y_pred_train = ANN_class.predict(X_train)
    y_pred_test = ANN_class.predict(X_test)
    acc_train_func.append(accuracy_score(y_train,y_pred_train))
    acc_test_func.append(accuracy_score(y_test,y_pred_test))
    
## Plotting the graph for the different hidden layer neurons 
plt.plot(function,acc_train_func,marker='o',markersize=6,color='blue',label='training')
plt.plot(function,acc_test_func,marker='o',markersize=6,color='red',label='test')
plt.title(" Accuracies for different Activations specified")
plt.xlabel("Activation functions")
plt.grid()
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```

```python
###Parameter3:Solver Function
types = ['lbfgs', 'sgd', 'adam']
acc_train_type = []
acc_test_type = []
for i in types:
    ANN_class = MLPClassifier(hidden_layer_sizes= (30,30,30,30),
                          activation= 'relu' , 
                          solver = i ,
                          random_state= 0 )
    ANN_class.fit(X_train, y_train)
    y_pred_train = ANN_class.predict(X_train)
    y_pred_test = ANN_class.predict(X_test)
    acc_train_type.append(accuracy_score(y_train,y_pred_train))
    acc_test_type.append(accuracy_score(y_test,y_pred_test))
    
## Plotting the graph for the different hidden layer neurons 
plt.plot(types,acc_train_type,marker='o',markersize=6,color='Blue',label='training')
plt.plot(types,acc_test_type,marker='o',markersize=6,color='red',label='test')
plt.title(" Accuracies for different Solvers specified")
plt.xlabel("Solver")
plt.grid()
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```

```python
## Parameter4:Alpha Values
values = [0.0001,0.0002,0.0004,0.0006,0.0008,0.001]
acc_train_values = []
acc_test_values = []
for i in values:
    ANN_class = MLPClassifier(hidden_layer_sizes= (30,30,30,30),
                          activation= 'relu' , 
                          alpha = i,solver= 'sgd' 
                          random_state= 0 )
    ANN_class.fit(X_train, y_train)
    y_pred_train = ANN_class.predict(X_train)
    y_pred_test = ANN_class.predict(X_test) 
    acc_train_values.append(accuracy_score(y_train,y_pred_train))
    acc_test_values.append(accuracy_score(y_test,y_pred_test))
    
## Plotting the graph for the different hidden layer neurons 
plt.plot(values,acc_train_values,marker='o',markersize=6,color='blue',label='training')
plt.plot(values,acc_test_values,marker='o',markersize=6,color='red',label='test')
plt.title(" Accuracies for different Alpha specified")
plt.xlabel("Alpha values")
plt.grid()
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```

```python
#####Final ANN Classifier

ANN_class= MLPClassifier(hidden_layer_sizes= (30,30,30,30),
                          activation='relu',  
                          batch_size='auto' ,
                          solver= 'sgd',
                          learning_rate= 'constant',
                          max_iter= 500,
                          random_state= 0 )
ANN_class.fit(X_train, y_train)

###Predicting
y_pred= ANN_class.predict(X_test)
y_expect = y_test

##Accuracy
acc_dt1 = accuracy_score(y_expect , y_pred)
print(acc_dt1)

### create confusion matrix
cm_dt1= confusion_matrix(y_test , y_pred)
print(cm_dt1)
```

```python
####Dataset2

df_data2=pd.read_csv("C:/Users/PC/Desktop/ML/Bank_Data.csv")
df_data2.head()

##Converting Variables to Categorical Data.

df_data2['default']=df_data2['default'].map({'unknown':0,'no':1,'yes':2})
df_data2['housing']=df_data2['housing'].map({'unknown':0,'no':1,'yes':2})
df_data2['loan']=df_data2['loan'].map({'unknown':0,'no':1,'yes':2})
df_data2['month']=df_data2['month'].map({'jan':1,'feb':2,'mar':3,'apr':4,'may':5,'jun':6,'jul':7,'aug':8,'sep':9,'oct':10,'nov':11,'dec':12})
df_data2['poutcome']=df_data2['poutcome'].map({'nonexistent':0,'failure':1,'success':2})
df_data=df_data2.drop(['duration'],axis=1)
#df_data2.head()

###Converting Nominal Variable.

collst=['job','marital','education','contact','day_of_week']
df_data2=pd.get_dummies(df_data2,columns=collst)

df_data2['y']=df_data2['y'].map({'yes':1,'no':0})

####Splitting Data into Train Test
X2=df_data2.drop('y',axis=1).values
Y2=df_data2['y'].values

####Splitting data into train test
X_train2,X_test2, y_train2, y_test2 = train_test_split(X2,Y2,test_size = 0.25 , random_state = 0)

```

```python
# Varying the Neighbor values in the classifier for p=2
n_neighbors = np.arange(1,21,1)
acc_train_neighbor = []
acc_test_neighbor = []
for i in n_neighbors:
    KNN_class = KNeighborsClassifier(n_neighbors=i , p = 2)
    KNN_class.fit(X_train, y_train)
    y_pred_train = KNN_class.predict(X_train)
    y_pred_test = KNN_class.predict(X_test)
    acc_train_neighbor.append(accuracy_score(y_train,y_pred_train))
    acc_test_neighbor.append(accuracy_score(y_test,y_pred_test))
        
## Plotting the graph for the different Neighbors with p=2 
plt.plot(n_neighbors,acc_train_neighbor,marker='o',markersize=6,color='blue',label='training')
plt.plot(n_neighbors,acc_test_neighbor,marker='o',markersize=6,color='red',label='test')
plt.title(" Accuracies for different Neighbors specified")
plt.xlabel("Number of Neighbors(k)")
plt.grid()
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```

```python
#####Checking for Parameter3:Weight
weight = ['uniform','distance']
acc_train_weight = []
acc_test_weight = []
for i in weight:
    KNN_class = KNeighborsClassifier(n_neighbors=15 , p = 2 , weights= i )
    KNN_class.fit(X_train, y_train)
    y_pred_train = KNN_class.predict(X_train)
    y_pred_test = KNN_class.predict(X_test)
    acc_train_weight.append(accuracy_score(y_train,y_pred_train))
    acc_test_weight.append(accuracy_score(y_test,y_pred_test))
    
## Plotting the graph for the different Neighbors with p=2 
plt.plot(weight,acc_train_weight,marker='o',markersize=6,color='blue',label='training')
plt.plot(weight,acc_test_weight,marker='o',markersize=6,color='red',label='test')
plt.title(" Accuracies for different Weights specified")
plt.xlabel("Weights - Uniform or Distance")
plt.grid()
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```

```python
###Parameter4: Algorithm
algo = ['auto','ball_tree','kd_tree','brute']
acc_train_algo = []
acc_test_algo = []
for i in algo:
    KNN_class = KNeighborsClassifier( algorithm= i  )
    KNN_class.fit(X_train, y_train)
    y_pred_train = KNN_class.predict(X_train)
    y_pred_test = KNN_class.predict(X_test)
    acc_train_algo.append(accuracy_score(y_train,y_pred_train))
    acc_test_algo.append(accuracy_score(y_test,y_pred_test))
    
## Plotting the graph for the different Neighbors with p=2 
plt.plot(algo,acc_train_algo,marker='o',markersize=6,color='blue',label='training')
plt.plot(algo,acc_test_algo,marker='o',markersize=6,color='red',label='test')
plt.title(" Accuracies for different Algorithms specified")
plt.xlabel("Algorithms")
plt.grid()
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```

```python
#####Final K_NN Classifier

KNN_class = KNeighborsClassifier(n_neighbors=6 , p = 3 , metric= 'minkowski' )
KNN_class.fit(X_train, y_train)

##Making Prediction
y_pred = KNN_class.predict(X_test)
y_expect = y_test


##Accuracy
acc_dt1 = accuracy_score(y_expect , y_pred)
print(acc_dt1)

### create confusion matrix
cm_dt1= confusion_matrix(y_test , y_pred)
print(cm_dt1)
```
